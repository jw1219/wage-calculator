```python
import ipywidgets as widgets

def minutes_to_hours(minutes):
  return minutes / 60

def calc_wage(hourly_wage, hours):
  return hourly_wage * hours

label_header = widgets.Label(value='Wage Calculator', layout=widgets.Layout(display='flex', justify_content='center'))
input_wage_hourly = widgets.BoundedIntText(
    value='2000',
    min=800,
    max=10000,
    step=10,
    description='Hourly Wage (yen):',
    disabled=False,
    style={'description_width': 'initial'}
)
input_hours = widgets.Text(
    placeholder='e.g. 3.5',
    description='Hours:',
    disabled=False
)
input_minutes = widgets.Text(
    placeholder='e.g. 45',
    description='Minutes:',
    disabled=False
)
btn_run = widgets.Button(description='Calculate', layout=widgets.Layout(height='auto', width='auto'), button_style='primary')
label_wage = widgets.Label(value='Your Wage (yen):')

def run(button):
  try:
    hourly_wage = int(input_wage_hourly.value)
  except:
    print('Invalid hourly wage')
    return
  try:
    hours = float(input_hours.value)
  except:
    print('Invalid hours')
    return
  try:
    minutes = float(input_minutes.value)
  except:
    minutes = 0
  total_hours = hours + minutes_to_hours(minutes)
  label_wage.value = f'Your Wage (yen): {calc_wage(int(hourly_wage), total_hours):.02f}'

btn_run.on_click(run)

grid = widgets.GridspecLayout(4, 2, grid_gap='10px')
grid[0, :] = input_wage_hourly
grid[1, 0] = input_hours
grid[1, 1] = input_minutes
grid[2, :] = btn_run
grid[3, :] = label_wage

widgets.AppLayout(
    header=label_header,
    left_sidebar=None,
    center=grid,
    right_sidebar=None,
    footer=None,
    grid_gap='10px'
)
```
